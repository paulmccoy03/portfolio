```python
# Question 1
import pandas as pd #Installing inital packages
import yfinance as yf
import numpy as np
```

```python
#Base NVIDIA data
tick= "NVDA"
yf.download(tick)
nvda=yf.download(tick)
NVDAdata= yf.download(tick, start= "2001-01-01", end= "2023-12-31")
```

```python
#Base Nasdaq data
NASDAQdata= yf.download("^IXIC", start="2001-01-01", end="2023-12-31")
```

```python
#Get Log Returns
NVDAdata["logreturn"]=np.log(NVDAdata["Adj Close"]) - np.log(NVDAdata["Adj Close"].shift(1))
NASDAQdata["logreturn"]=np.log(NASDAQdata["Adj Close"]) - np.log(NASDAQdata["Adj Close"].shift(1))
```

```python
#Cleaning data
nvda=NVDAdata.dropna()
nasdaq=NASDAQdata.dropna()
```

```python
#Question 3
from matplotlib import pyplot as plt
plt.figure(figsize=(16,10), dpi=150) 
nvda["logreturn"].plot(label="NVDA", color="red")
nasdaq["logreturn"].plot(label="NASDAQ", color="black")
plt.title("Daily Return for NVDA and NASDAQ from 1/1/01 to 12/31/23")
plt.xlabel("Years")
plt.ylabel("Percent Return")
plt.legend(fontsize=14)
#This is right but super small
```

```python
#Same as data above, just monthly instead of daily for easier analysis
nvda_monthly = nvda["logreturn"].resample("M").sum()
nasdaq_monthly = nasdaq["logreturn"].resample("M").sum()
fig, ax = plt.subplots(figsize=(16, 8), dpi=450)
nvda_monthly.plot(ax=ax, label="NVDA", color="red")
nasdaq_monthly.plot(ax=ax, label="NASDAQ", color="black")
ax.set_title("Monthly Return for NVDA and NASDAQ from 1/1/01 to 12/31/23", fontsize=16)
ax.set_xlabel("Years", fontsize=14)
ax.set_ylabel("Percent Return", fontsize=14)
ax.legend(fontsize=14)
plt.show()
```

```python
#NVIDIA shows way more variation within the data
#They both follow the same general trend for the most part, but NVIDIA shows the same movement to a way greater magnitude.
#To call it a true pattern is a stretch, but there may be slight positve correlation.
```

```python
#Question 5
cor=np.corrcoef(nvda["logreturn"], nasdaq["logreturn"])[0,1]
cor
```

```python
#significance (Hypothesis) testing
# H0: B=0    ???**
# Ha: B does not=0    ???**
# DR: Reject the null (H0) if |t| > t_sub_alpha/2, n-2
```

```python
#T-stat
from scipy import stats
#equal var because 
t_statistic, p_value_equalvar=stats.ttest_ind(nasdaq["logreturn"], nvda["logreturn"],equal_var=True, alternative="two-sided")
print(t_statistic, p_value_equalvar)
```

```python
#Decesion
def decision_scipy_ttest(pvalue, alpha):
 if pvalue < alpha:
    print(f"Reject the null at {alpha} level of significance.")
 else:
    print(f"Accept the null at {alpha} level of significance.")
decision_scipy_ttest(p_value_equalvar,.05)
```

```python
#The results in the previous cells formally show that theres statistically significant evidence that relation shown in covar and correlation are right, and that there is postive linear realtionship
```

```python
#Question 6
import statsmodels.api as sm
x=nasdaq["logreturn"]
y=nvda["logreturn"]
x_sm=sm.add_constant(x)
ols=sm.OLS(y,x_sm).fit()
print(ols.summary())
```

```python
#The previous regression shows a constant coeffcient of .0005, which is basically zero. This can be interpreted as when the NASDAQ has returns close to 0 (i.e. no movement),NVDA will also have returns close to 0
#The log return coeffcient is 1.5583, which shows that NVDA's returns are sensitive to the NASDAQ. As the value is greater than one, we can also predict that the beta of NVDA will be greater than the market, hinting at comparibly higher volitilty.
# All this to say that the two stocks are lineraly postivly related, and that NVDA is more volitile than the examined benchmark.
```
